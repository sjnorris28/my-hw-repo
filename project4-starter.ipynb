### Intro: Write a problem Statement/ Specific Aim for this project

Answer: What is the relationship between graduate school admission and the prestige (4 rankings possible) of the undergraduate institution where the applicant attended.

### Dataset:  Write up a description of your data and any cleaning that was completed

Answer: Aside from prestige, which had four possible rankings, GRE and GPA were also variables in the data, which ranged from 200-800 and 0-4, respectively.

### Methods: Write up the methods used in your analysis

Answer: We calculated the probability of admissions to graduate school based on the prestige of the undergraduate school they attended, controlling for GPA and GRE.   

### Results: Write up your results

Answer: When GRE and GPA are controlled, a student from a prestige ranked school of 1 had a 73% chance of admission, prestige rank of 2 had 58%, rank of 3 had 42%, and rank of 4 had 37% chance.
